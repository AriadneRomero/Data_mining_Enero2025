```python
# Ariadne Selena Romero Rivero #1957540
from pandasql import sqldf
from sqlalchemy import engine

# Writing functions in SQL
pysqldf = lambda q: sqldf(q, globals())
```

```python
import pandas as pd
import numpy as np
data = pd.read_csv("videogamesales/videogamesales_clean.csv")
data
```

```python
# Columns
data.info()
```

```python
# Estadisticas Generales: General Statistics
data.describe()
```

## Analysis of Sales

```python
# Ventas por gÃ©nero de juego con .describe
data.groupby('Genre')['Global_Sales'].describe()
```

```python
# Ventas por gÃ©nero de juego por millon
ventas_genero = data.groupby('Genre')['Global_Sales'].aggregate(['count', 'mean', 'sum', 'min', 'max']).sort_values(by='sum', ascending=False)
ventas_genero.columns = ['Total_Juegos', 'Promedio_Ventas', 'Ventas_Totales (millones)', 'Venta_MÃ­nima', 'Venta_MÃ¡xima']
ventas_genero
```

```python
# Ventas por gÃ©nero de juego con Sintaxis SQL
q = """ 
SELECT Genre, COUNT(Global_Sales), AVG(Global_Sales), MAX(Global_Sales), MIN(Global_Sales) 
FROM data
GROUP BY Genre
"""
pysqldf(q)
```

```python
# Ventas por plataforma por millon
ventas_plataforma = data.groupby('Platform')['Global_Sales'].aggregate(['count', 'mean', 'sum', 'min', 'max']).sort_values(by='sum', ascending=False)
ventas_plataforma.columns = ['Total_Juegos', 'Promedio_Ventas', 'Ventas_Totales(millones)', 'Venta_MÃ­nima', 'Venta_MÃ¡xima']
ventas_plataforma.head(10)
```

```python
# Ventas por marca
ventas_publi = data.groupby('Publisher')['Global_Sales'].aggregate(['count', 'mean', 'sum', 'min', 'max']).sort_values(by='sum', ascending=False)
ventas_publi.columns = ['Total_Juegos', 'Promedio_Ventas', 'Ventas_Totales(millones)', 'Venta_MÃ­nima', 'Venta_MÃ¡xima']
ventas_publi.head(20)
```

```python
venta_regiones = data[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()
venta_regiones
```

## Analysis of years

## Analysis of trend

```python
# NÃºmero de gÃ©neros
popular_genre = data['Genre'].value_counts()
popular_genre
```

```python
# Top 15 de videojuegos mÃ¡s comprados
top_15 = data.groupby(['Name', 'Platform'])['Global_Sales'].aggregate(['max']).sort_values(by='max', ascending=False)
top_15.columns = ['Venta']
top_15.head(15)
```
