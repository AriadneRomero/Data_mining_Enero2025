```python
# Ariadne Selena Romero Rivero #1957540
import pandas as pd
import numpy as np

import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
data = pd.read_csv("../videogamesales/videogamesales_clean.csv")
data
```

```python
genre_counts = data['Genre'].value_counts()
genre_counts
```

**HipÃ³tesis nula (Hâ‚€):** No hay diferencia en las ventas medias entre gÃ©neros 
\
**HipÃ³tesis alternativa (Hâ‚):** Hay diferencias significativas entre gÃ©neros

```python
# Modelo ANOVA
modl = ols('Global_Sales ~ Genre', data=data).fit()
anova_df = sm.stats.anova_lm(modl, typ=2)

valor_p = anova_df["PR(>F)"].iloc[0]

if valor_p < 0.05:
    print("âœ”ï¸ Hay diferencias significativas entre gÃ©neros")
    print(anova_df)

    # Prueba post-hoc de Tukey
    tukey = pairwise_tukeyhsd(endog=data["Global_Sales"],
                                groups=data["Genre"],
                                alpha=0.05)

    print("\nðŸ” Resultados de la prueba de Tukey:")
    print(tukey)

else:
    print("âŒ No hay diferencias significativas entre gÃ©neros")
```
